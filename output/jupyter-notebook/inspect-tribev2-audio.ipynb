# Inspect TRIBE v2 Audio Predictions

Use this notebook after running `scripts/test_tribev2_raw_audio.py`.
It loads the saved prediction tensor and basic metadata for quick inspection.


```python
from __future__ import annotations

import json
from pathlib import Path

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

OUTPUT_DIR = Path("output/latest")
OUTPUT_DIR
```

```python
summary = json.loads((OUTPUT_DIR / "summary.json").read_text())
segments = pd.read_csv(OUTPUT_DIR / "segments.csv")
events = pd.read_csv(OUTPUT_DIR / "events.csv")
preds = np.load(OUTPUT_DIR / "predictions.npy")

summary
```

```python
print("predictions shape:", preds.shape)
print("dtype:", preds.dtype)
print("min/max:", float(preds.min()), float(preds.max()))
print("mean/std:", float(preds.mean()), float(preds.std()))

segments
```

```python
events
```

```python
segment_means = preds.mean(axis=1)
segment_stds = preds.std(axis=1)

pd.DataFrame(
    {
        "segment_index": np.arange(len(segment_means)),
        "mean": segment_means,
        "std": segment_stds,
    }
)
```

```python
plt.figure(figsize=(10, 4))
plt.plot(segment_means, marker="o")
plt.title("Mean predicted response per kept segment")
plt.xlabel("Segment index")
plt.ylabel("Mean response")
plt.grid(True, alpha=0.3)
plt.show()
```

```python
vertex_slice = preds[:, :256]
plt.figure(figsize=(12, 4))
plt.imshow(vertex_slice, aspect="auto", cmap="viridis")
plt.colorbar(label="Predicted response")
plt.title("Prediction heatmap for the first 256 vertices")
plt.xlabel("Vertex index")
plt.ylabel("Segment index")
plt.show()
```
